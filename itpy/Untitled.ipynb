```python
from re import match

def print_code(module, rturn):

    with open("{}.py".format(module)) as fn:
        lines = fn.readlines()

    def deco(s):
        m = match("[a-zA-z\s]+", s)
        fname = m.group().split(" ")[-1]
        return "@iter_wraps({}.{})\n".format(module, fname)
    
    for line in lines:
        if line.startswith("def"):
            print(deco(line) + line + "\treturn {}".format(rturn), "\n")
```

```python
print("# transforms")
print_code("transforms", "Itpy(iter=iterable)")
print("# summary")
print_code("summary", "Itpy.VALUE")
print("# iio")
print_code("iio", "Itpy()")
```
